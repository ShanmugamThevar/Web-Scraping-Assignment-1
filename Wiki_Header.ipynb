# Wiki MainPage Header

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
data=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
data
```

```python
data.content
```

```python
soup=BeautifulSoup(data.content, 'html.parser')
```

```python
header_tags=[]
```

```python
header=soup.find_all('span', class_="mw-headline")
header[:]
```

```python
for i in header:
    header_tags.append(i.get_text())
header_tags[:]
```

```python
import pandas as pd
Wiki=pd.DataFrame({})
Wiki['Header']=header_tags[0:8]
```
